# 1. Private Construction spending data
### As this assignment requires the Error named R2x need to be showed on the dataframe, so there is a new columns R2Private which is the the error for private construction costs data. 

### The column right next the R2Private is R2ErrorPrivate which is the ratio of the difference between private construction costs and the moving average,divided by the private construction costs.

### The squared error column named sqErr, which means the average squared difference between the estimated values and the actual value, and column sqErr is on the left side of R2Private column. 


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv('C:/Users/evgpr/OneDrive/Desktop/WMPython/CTBA/M3/ConstructionTimeSeriesDataV2.csv')

def sqErr(row):
    return (row['Private Construction'] - row['MovAvg'])**2
    
df['MovAvg'] = np.NaN
df['sqErr'] = np.NaN

df.index = pd.DatetimeIndex(freq='m',data=pd.date_range(freq='m',start=pd.Timestamp(year=2002,month=1,day=31),periods=len(df['Private Construction'])))

window = 36
window = window - window % 2

for i in range(int(window/2),df.shape[0]-int(window/2)):
    df.loc[df.index[i],'MovAvg'] = (0.5*df.iloc[i - int(window/2)]['Private Construction'] + df.iloc[i - int(window/2)+1:i + int(window/2)]['Private Construction'].sum() + 0.5*df.iloc[i + int(window/2)]['Private Construction'])/float(window)

    df['sqErr'] = (df['Private Construction'] - df['MovAvg'])**2
    
df.dropna(how='any',inplace=True)

df['R2Private'] = df['Private Construction'] - df['MovAvg']
df['R2ErrorPrivate'] = abs((df['Private Construction'] - df['R2Private'])/df['Private Construction'])

df.style.format({
    'MovAvg': '{:.1f}'.format,
    'sqErr': '{:,.1f}'.format,
    'R2Private': '{:,.1f}'.format,
    'R2ErrorPrivate': '{:,.3f}'.format
})

pd.set_option('display.max_rows',200)
pd.set_option('display.max_columns',200)
pd.set_option('display.width',200)

print(df)

```
